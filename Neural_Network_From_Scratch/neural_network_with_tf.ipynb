```python
import tensorflow as tf
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.activations import relu,linear
from tensorflow.keras.losses import SparseCategoricalCrossentropy
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.datasets import mnist
from sklearn.model_selection import train_test_split
```

```python
#Load and preprocess dataset
(X_train, Y_train), (X_, y_) = mnist.load_data()
```

```python
X_cv, X_test, y_cv, y_test = train_test_split(X_,y_,test_size=0.20, random_state=1)
print("X_train.shape", X_train.shape, "y_train.shape", y_train.shape)
print("X_cv.shape", X_cv.shape, "y_cv.shape", y_cv.shape)
print("X_test.shape", X_test.shape, "y_test.shape", y_test.shape)
```

```python
#preprocess dataset
X_train = X_train.reshape(X_train.shape[0],28*28).astype('float32')/255
X_test = X_test.reshape(X_test.shape[0],28*28).astype('float32')/255
X_cv = X_cv.reshape(X_cv.shape[0],28*28).astype('float32')/255

print("X_train.shape", X_train.shape, "y_train.shape", y_train.shape)
print("X_cv.shape", X_cv.shape, "y_cv.shape", y_cv.shape)
print("X_test.shape", X_test.shape, "y_test.shape", y_test.shape)
```

```python
tf.random.set_seed(1234)
model = Sequential(
    [
        
        Dense(units=25,activation="relu",kernel_regularizer=tf.keras.regularizers.l2(0.01)),
        Dense(units=15,activation="relu",kernel_regularizer=tf.keras.regularizers.l2(0.01)),
        Dense(units=10,activation="linear")
        
        
        
        
              ], name= None
)
model.compile(
   
    loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
    metrics=['accuracy'],
    optimizer=tf.keras.optimizers.Adam(0.001),
     
)
```

```python
model.summary()
```

```python
model.fit(
    X_train, y_train,
    epochs=20,
    batch_size=32
)
```

```python
logits = model.predict(X_test)
probabilities = tf.nn.softmax(logits).numpy()
print(probabilities.s
```

```python
y_pred = np.argmax(probabilities, axis=1)
accuracy = np.mean(y_pred == y_test)
print("Test accuracy:", accuracy)
print(y_pred[0])
print(y_test[0])
```
